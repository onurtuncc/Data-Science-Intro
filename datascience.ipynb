### Soru 1:
* Stackoverflow'un 2019 yılı için geliştiricilerle yaptığı anket sonuçlarını içeren ```"survey_results_public.csv"``` adlı dosyayı okuyarak __survey_19__ adlı DataFrame'e aktarın.
* DataFrame'in satır ve sütun sayısını yazdırın.
* ```Respondent``` sütununu index olarak atayın.
* Aşağıdaki sütunları dataframe'den siliniz:

    ``` ['ScreenName', 'SOVisit1st','SOVisitFreq', 'SOVisitTo', 'SOFindAnswer', 'SOTimeSaved', 'SOHowMuchTime', 'SOAccount',   'SOPartFreq', 'SOJobs', 'EntTeams', 'SOComm', 'WelcomeChange', 'SONewContent']``` 
  
  
* ```DevType``` ya da ```LanguageWorkedWith``` sütununda eksik veri olan satırları siliniz. Kaç satır silindi?
* ```Country``` sütunundaki ```Russian Federation``` ve ```Czech Republic``` değerlerini sırasıyla ```Russia``` ve ```Czechia``` olarak değiştiriniz.
* Haftalık çalışma saati ```(WorkWeekHrs)``` ya da yıllık maaş ```(ConvertedComp)``` sütunlarındaki değerleri, ilgili sütunun ortalamasının %99'undan daha uzakta veri içeren satırları siliniz.
* ```Gender``` sütununda ```Man```, ```Woman``` ve ```NaN``` dışındaki değerleri, ```Non-binary``` değeri ile değiştirin.

````python
import pandas as pd
````

````python
survey_19=pd.read_csv("survey_results_public.csv")
print("Satır sayısı: ",len(survey_19))
print("Sütun sayısı: ",len(survey_19.columns))
survey_19
````

````python
survey_19=survey_19.set_index("Respondent")
survey_19
````

````python
survey_19.drop(columns=['ScreenName', 'SOVisit1st','SOVisitFreq', 'SOVisitTo', 'SOFindAnswer',
                        'SOTimeSaved', 'SOHowMuchTime', 'SOAccount',   'SOPartFreq', 'SOJobs',
                        'EntTeams', 'SOComm', 'WelcomeChange', 'SONewContent'],inplace=True)
````

````python
sayi1=len(survey_19)
nan_value = float("NaN")
survey_19.replace("", nan_value, inplace=True)
survey_19.dropna(subset=["DevType","LanguageWorkedWith"], inplace=True)
print("Silinen satır sayısı = ",sayi1-len(survey_19))
````

````python
survey_19["Country"].replace(to_replace=["Russian Federation","Czech Republic"],value=["Russia","Czechia"], inplace=True)
survey_19["Country"]
````

````python
wwh_average= survey_19['WorkWeekHrs'].mean()
cc_average= survey_19['ConvertedComp'].mean()
survey_19.drop(survey_19['ConvertedComp'][(survey_19['ConvertedComp']-cc_average>cc_average*99/100) |((cc_average-(cc_average*99/100))>survey_19['ConvertedComp'])].index,inplace=True)
survey_19.drop(survey_19['WorkWeekHrs'][(survey_19['WorkWeekHrs']-wwh_average>wwh_average*99/100)|((wwh_average-(wwh_average*99/100))>survey_19['WorkWeekHrs'])].index,inplace=True)
````

````python
survey_19['Gender'] = survey_19['Gender'].apply(lambda x: "Non-binary" if ((pd.notna(x))&(x!='Man')&(x!='Woman')) else x)
survey_19["Gender"].value_counts()
````

### Soru 2:
* ```Age``` sütununu 0, 15, 19, 24, 29, 34, 39, 44, 49, 54, 59, 99 binleri ile aralığa çevirerek ```AgeInterval``` adında yeni bir sütun oluşturunuz.
* Ankete katılanların yaş aralıklarına göre dağılımını pasta grafiğinde gösteriniz. 
* (19-24] yaş aralığının en çok kullandığı sosyal medya nedir?

````python
yas_araliklar=["(0-15]","(15-19]","(19-24]","(24-29]","(29-34]","(34-39]","(39-44]","(44-49]","(49-54]","(54-59]","(59,99000]"]
survey_19['AgeInterval'] = survey_19["Age"].apply(lambda row: yas_araliklar[0] if ((row>=0)&(row<=15)) else(yas_araliklar[1] if(row<=19) else(yas_araliklar[2] if(row<=24) else(yas_araliklar[3] if(row<=29) else(yas_araliklar[4] if(row<=34) else(yas_araliklar[5] if(row<=39) else(yas_araliklar[6] if(row<=44) else(yas_araliklar[7] if(row<=49) else(yas_araliklar[8] if(row<=54) else(yas_araliklar[9] if(row<=59) else (yas_araliklar[10] if row<=99000 else nan_value)))))))))))
survey_19[["Age","AgeInterval"]]

````

````python
df=pd.DataFrame({"Yas aralıklarına gore dagılım":survey_19["AgeInterval"].value_counts()},index=yas_araliklar)
plot=df.plot.pie(y="Yas aralıklarına gore dagılım",figsize=(8,8))
````

````python
sos_med=survey_19[survey_19["AgeInterval"]=="(19-24]"]
print("19-24 yaş aralığının en çok kullandığı sosyal medya : ",sos_med["SocialMedia"].mode()[0])

````

### Soru 4: 
Kadınların daha çok tercih ettiği/edildiği (ankete katılan kadın-erkek oranından daha yüksek orana sahip) geliştirici tipleri ```(DevType)```nelerdir?
* Her bir ```DevType``` tipinin sütun olarak, ankete katılanların da satır olarak temsil edildiği yeni bir DataFrame oluşturup __devtype_df__ adlı değişkene atayın.
* Oluşan __devtype_df__ DataFrame'ine ```Gender``` sütununu ekleyin.
* DataFrame'i cinsiyete göre gruplayarak her bir ```DevType``` için cinsiyete göre sayıları bulun.
* ```DevType``` tiplerindeki kadın-erkek oranlarını, ankete katılan kadın-erkek oranı ile karşılaştırarak kadınların daha çok tercih ettiği/edildiği ```DevType``` tiplerini listeleyin.

````python
k_liste=[]
d=survey_19["DevType"].str.split(";",expand=True)
cols=d[0].unique()
t=list(map(lambda x:survey_19["DevType"].str.contains(x),cols))
dd=pd.DataFrame(t)
df=dd.T
df.columns=cols
devtype_df=df
devtype_df["Gender"]=survey_19["Gender"]
g=devtype_df.groupby("Gender")
ake_oran=survey_19["Gender"].value_counts()[1]/survey_19["Gender"].value_counts()[0]
cc=list(map(lambda x:g.get_group("Woman")[x].value_counts()[1]/g.get_group("Man")[x].value_counts()[1],cols))
dff=pd.DataFrame(cc)
k_dff=dff[dff[0]>ake_oran]
k_liste=list(map(lambda x :cols[k_dff.index[x]],range(len(k_dff))))
print("Kadınların daha çok tercih edildiği DevType tipleri :",k_liste)
df
````

### Soru 13:
* https://download.geonames.org/export/dump/countryInfo.txt adresindeki veriyi:
    * ilk 49 satırını ihmal ederek
    * yalnızca ```Country, Continent, CurrencyCode``` sütunlarını kullanarak 
    * ```Country``` sütununu indeks sütun olarak 
    * ```NA``` filtreleme yapmadan (```North America``` kıtasının kısaltması olan ```NA```'yı ```NaN``` değer olarak okumamak için)
  
  okuyarak __country_info__ adlı DataFrame'e aktarınız.
  
  
*     

           {AF : Africa
            AS : Asia			
            EU : Europe			
            NA : North America		
            OC : Oceania			
            SA : South America		
            AN : Antarctica	} 

   yukarıdaki haritalamayı kullanarak kıta isimlerini içeren ```ContinentName``` isimli yeni bir sütun oluşturunuz.

````python
country_info=pd.read_csv("countryInfo.txt",skiprows=49,usecols=["Country","Continent","CurrencyCode"],
                         sep="\t",keep_default_na=False,index_col="Country")

country_info
````

````python
continent_names={"AF" : "Africa",
  "AS" : "Asia",            
  "EU" : "Europe"    ,        
  "NA" : "North America" ,       
  "OC" : "Oceania"     ,       
  "SA" : "South America"     ,   
  "AN" : "Antarctica"    }
country_info["ContinentName"]=country_info["Continent"].apply(lambda x:continent_names[x])
country_info
````

### Soru 14:

* __survey_19__ dataframe'i ile __country_info__ dataframe'ini ülke bazında birleştirip __merged_survey__ isimli yeni bir dataframe'e aktarınız.
* Yaşadığı ülkenin resmi para biriminin dışındaki para birimiyle maaş alan kişi sayısı nedir?
